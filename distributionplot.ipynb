```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
word2id={}
word2id['<PAD>'] = 0
word2id['<S>'] = 1
word2id['</S>'] = 2
word2id['<UNK>'] = 3
word2id['0'] = 4
word2id['4'] = 5
word2id['5'] = 6
word2id['1'] = 7
word2id['6'] = 8
word2id['3'] = 9
word2id['7'] = 10
word2id['2'] = 11
word2id['d'] = 12
word2id['9'] = 13
word2id['8'] = 14
word2id['c'] = 15
word2id['a'] = 16
word2id['e'] = 17
word2id['f'] = 18
word2id['b'] = 19


def sentence_to_ids(word2id, sentence, UNK=3):
    ids = [word2id.get(word, UNK) for word in sentence]
    return ids

def load_data(file_path):
    data = []
    for line in open(file_path, encoding='utf-8'):
        words = line.strip().split()
        data.append(words)
    return data

data_pos=load_data("data/train/emqx_whole_len50.txt")
data_neg=load_data("data/save/test/emqx_whole_len50_epoch_1_test.txt")
dataids_pos=[]
dataids_neg=[]

for d in data_pos:
    dataids_pos.append(sentence_to_ids(word2id,d))
    
for d in data_neg:
    dataids_neg.append(sentence_to_ids(word2id,d))
```

Training for 75 epochs(epoch_1)

```python
fig = plt.figure()
for i in range(1):
    sns.kdeplot(dataids_pos[2],color='red',legend=False)
    sns.kdeplot(dataids_neg[13],color='blue',legend=False)
plt.legend(['real','synthetic'],loc='upper right')
fig.savefig("epoch_75_train.pdf")
```
